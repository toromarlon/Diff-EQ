**Predator-prey System**

If no lynx are present the hares will grow exponentially $\frac{dH}{dt} = aH$

With lynx present, the rate of growth becomes $\frac{dH}{dt} = aH - bHL$

With no food present, the lynx population will decline at $\frac{dL}{dt} = -cL$

With hares present, the lynx population will grow at $\frac{dL}{dt} = -cL - dHL$

Now we have a **system** of equations where both interact:$\frac{dH}{dt} = aH - bHL$ and $\frac{dL}{dt} = -cL - dHL$F
