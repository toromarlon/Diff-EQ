**Spring-mass Model**

The following equations are used when measuring spring forces

Hooke's Law: $F = -kx$

$F(x) = F(0) + F^\prime(0)x + \frac{1}{2}F^{\prime \prime}(0)x^2 + ... = -kx + \frac{1}{2}F^{\prime \prime}(0)x^2 + ....,$

Newton's 2nd Law of Motion: $F = ma = m(d^2x/dt^2) = mx^{\prime \prime}$



We get second order differential equation: $mx^{\prime \prime} = -kx$


Adding a dampener to our system we add the additional force $F = -bx^\prime$


Our new equation for our ***simple damped harmonic oscillator*** is now: $mx^{\prime \prime} + bx^{\prime} + kx$
